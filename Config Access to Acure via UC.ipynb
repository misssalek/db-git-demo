Create the storage connection to gizmobox containter on Azure

```sql
%sql
create external location if not exists dea_course_ext_dl_gizmobox
url 'abfss://gizmobox@daecourseexdl.dfs.core.windows.net/'
with (storage credential dea_course_sc)
```

Create Catalog (SQL) and Schemas in moedalion model ( Landing Gold Silver and Bronz)

```sql
%sql
create catalog if not exists gizmobox
managed location  'abfss://gizmobox@daecourseexdl.dfs.core.windows.net/'
```

```sql
USE CATALOG gizmobox;
SHOW SCHEMAS;
```

```sql
create schema if not exists landing
managed location 'abfss://gizmobox@daecourseexdl.dfs.core.windows.net/landing';

create schema if not exists gold
managed location 'abfss://gizmobox@daecourseexdl.dfs.core.windows.net/gold';

create schema if not exists silver
managed location 'abfss://gizmobox@daecourseexdl.dfs.core.windows.net/silver';

create schema if not exists bronze
managed location 'abfss://gizmobox@daecourseexdl.dfs.core.windows.net/bronze';
```

```sql
USE CATALOG gizmobox;
USE SCHEMA landing;
CREATE EXTERNAL VOLUME IF NOT EXISTS operational_data
  LOCATION 'abfss://gizmobox@daecourseexdl.dfs.core.windows.net/landing/operational_data/';
```
